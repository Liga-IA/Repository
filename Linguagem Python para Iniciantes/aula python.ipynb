# Topico 02: Operadores

#### Operadores Aritméticos:

- '+' (adição): soma dois valores.
- '-' (subtração): subtrai o segundo valor do primeiro valor.
- '*' (multiplicação): multiplica dois valores.
- '/' (divisão): divide o primeiro valor pelo segundo valor.
- '%' (módulo): retorna o resto da divisão do primeiro valor pelo segundo valor.
- '**' (exponenciação): calcula o primeiro valor elevado ao segundo valor.
- '//' (divisão inteira): retorna o resultado da divisão como um número inteiro.


```python
a = 10
b = 3

soma = a + b
print("Soma: ", soma)  # Resultado: 13

subtracao = a - b
print("Subtracao: ", subtracao)  # Resultado: 7

multiplicacao = a * b
print("Multiplicacao: ", multiplicacao)  # Resultado: 30

divisao = a / b
print("Divisao: ", divisao)  # Resultado: 3.3333333333333335

resto = a % b
print("Resto: ", resto)  # Resultado: 1

exponenciacao = a ** b
print("Exponenciacao: ", exponenciacao)  # Resultado: 1000

divisao_inteira = a // b
print("Divisao_inteira: ", divisao_inteira)  # Resultado: 3
```


#### Operadores de Atribuição:

- '=' (atribuição): atribui um valor à variável.
- '+=', '-=', '*=', '/=', '%=', '**=', '//=': operadores de atribuição combinados. Eles realizam a operação correspondente e atribuem o resultado à variável.


```python
x = 5
x += 3
print("Atribuição: ", x)  # Resultado: 8

y = 10
y -= 2
print("Atribuição: ", y)  # Resultado: 8

z = 7
z *= 2
print("Atribuição: ", z)  # Resultado: 14

w = 12
w /= 3
print("Atribuição: ", w)  # Resultado: 4.0
```


#### Operadores de Comparação:

- '==' (igual a): verifica se dois valores são iguais.
- '!=' (diferente de): verifica se dois valores são diferentes.
- '>' (maior que): verifica se o primeiro valor é maior que o segundo valor.
- '<' (menor que): verifica se o primeiro valor é menor que o segundo valor.
- '>=' (maior ou igual a): verifica se o primeiro valor é maior ou igual ao segundo valor.
- '<=' (menor ou igual a): verifica se o primeiro valor é menor ou igual ao segundo valor.



```python
a = 5
b = 3

igual = (a == b)
print("operador de igualdade: ", igual)  # Resultado: False

diferente = (a != b)
print("operador de diferenciação: ", diferente)  # Resultado: True

maior = (a > b)
print("operador de maior: ", maior)  # Resultado: True

menor = (a < b)
print("operador de menor: ", menor)  # Resultado: False

maior_igual = (a >= b)
print("operador de maior ou igual: ", maior_igual)  # Resultado: True

menor_igual = (a <= b)
print("operador de menor ou igual: ", menor_igual)  # Resultado: False
```


#### Operadores Lógicos:

- 'and' (e): retorna verdadeiro se ambas as condições forem verdadeiras.
- 'or' (ou): retorna verdadeiro se pelo menos uma das condições for verdadeira.
- 'not' (não): inverte o valor lógico.

```python
x = 5
y = 3
z = 7

resultado_and = (x > y) and (x < z)
print(resultado_and)  # Resultado: True

resultado_or = (x > y) or (x > z)
print(resultado_or)  # Resultado: True

resultado_not = not (x > y)
print(resultado_not)  # Resultado: False
```

### Estrutura de seleção

Existem tres estruturas de seleção principais em Python: o **if (se)** e o **if-else (se-senão)** e o **elif(se-senão-se)**


```python
idade = 25
altura = 180

# estrutura de condição IF combinada com operadores
if idade >= 18 and altura >= 170:
    print("Você é maior de idade e tem altura suficiente.")
```

```python
linguagem = "Python"

# estrutura de condição IF-ELSE
if linguagem == "Python":
    print("Linguagem de programação Python.")
else:
    print("Outra linguagem de programação.")
```

```python
numero = -3

# estrutura de condição ELIF
if numero > 0:
    print("O número é positivo.")
elif numero < 0:
    print("O número é negativo.")
else:
    print("O número é zero.")
```

# Topico 04: Estrutura de Repetição

- while
- for


### Estrutura de repetição

Existem duas estruturas de repetição principais em Python: **while (enquanto**) e **for (para)**. Aqui está uma explicação de cada uma delas:


```python
contador = 1

#  estrutura de repetição WHILE
while contador <= 5:
    print(contador)
    contador += 1
```

```python
frutas = ['maçã', 'banana', 'laranja']

# estrutura de repetição FOR
for fruta in frutas:
    print(fruta)
```

```python
contador = 0

# estrutura de repetição com condição de parada
while True:
    print(contador)
    contador += 1
    if contador == 5:
        break
```

```python
numeros = [1, 2, 3, 4, 5]

# estrutura de repetição com condição de continuidade
for numero in numeros:
    if numero == 3:
        continue
    print(numero)
```

# Topicos 05: Vetores

você pode criar vetores (também conhecidos como listas) para armazenar coleções de elementos. Os vetores podem conter diferentes tipos de dados, como números inteiros, números de ponto flutuante, strings e até mesmo outros vetores.

```python
# vetor ou lista
numeros = [1, 2, 3, 4, 5]
print(numeros[0])  # Saída: 1
print(numeros[2])  # Saída: 3

frutas = ["maçã", "banana", "laranja"]
print(frutas[1])  # Saída: banana

misto = [1, "dois", 3.0, [4, 5]]
print(misto[3])  # Saída: [4, 5]
```

```python
#  modificar um elemento no vetor

frutas = ["maçã", "banana", "laranja"]
frutas[1] = "abacate"
print(frutas)  # Saída: ["maçã", "abacate", "laranja"]
```

```python
# percorer um vetor com loop

frutas = ["maçã", "banana", "laranja"]
for fruta in frutas:
    print(fruta)
```

```python
#  a função len() retorna o tamanho (número de elementos) de um vetor.

numeros = [1, 2, 3, 4, 5]
tamanho = len(numeros)
print(tamanho)  # Saída: 5
```

```python
# a função append() é usada para adicionar um elemento ao final do vetor.

numeros = [1, 2, 3]
numeros.append(4)
print(numeros)  # Saída: [1, 2, 3, 4]
```

```python
# o método insert() é usado para adicionar um elemento em uma posição específica do vetor.

frutas = ["maçã", "banana", "laranja"]
frutas.insert(1, "abacate")
print(frutas)  # Saída: ["maçã", "abacate", "banana", "laranja"]
```

```python
# o método remove() é usado para remover a primeira ocorrência de um elemento específico do vetor.

numeros = [1, 2, 3, 2, 4]
numeros.remove(2)
print(numeros)  # Saída: [1, 3, 2, 4]
```

```python
# o método pop() é usado para remover e retornar o último elemento do vetor ou um elemento em uma posição específica.

numeros = [1, 2, 3, 4]
ultimo_elemento = numeros.pop()
print(ultimo_elemento)  # Saída: 4
print(numeros)  # Saída: [1, 2, 3]

numeros = [1, 2, 3, 4]
elemento = numeros.pop(1)
print(elemento)  # Saída: 2
print(numeros)  # Saída: [1, 3, 4]
```

```python
# o método sort() é usado para ordenar os elementos do vetor em ordem crescente.

numeros = [3, 1, 4, 2]
numeros.sort()
print(numeros)  # Saída: [1, 2, 3, 4]
```

```python
# o método reverse() é usado para inverter a ordem dos elementos do vetor.

numeros = [1, 2, 3, 4]
numeros.reverse()
print(numeros)  # Saída: [4, 3, 2, 1]
```

```python
# o operador in verifica se um elemento está presente no vetor.

frutas = ["maçã", "banana", "laranja"]
print("maçã" in frutas)  # Saída: True
print("abacaxi" in frutas)  # Saída: False
```

```python
# o método count() retorna o número de ocorrências de um elemento no vetor.

numeros = [1, 2, 3, 2, 4]
ocorrencias = numeros.count(2)
print(ocorrencias)  # Saída: 2
```

```python
# O método index() retorna o índice da primeira ocorrência de um elemento no vetor.

numeros = [1, 2, 3, 2, 4]
indice = numeros.index(2)
print(indice)  # Saída: 1
```

```python
# o método extend() é usado para adicionar vários elementos a um vetor existente.

numeros = [1, 2, 3]
outros_numeros = [4, 5, 6]
numeros.extend(outros_numeros)
print(numeros)  # Saída: [1, 2, 3, 4, 5, 6]
```

```python
# a função sorted() retorna uma nova lista ordenada com base nos elementos do vetor.

numeros = [3, 1, 4, 2]
numeros_ordenados = sorted(numeros)
print(numeros_ordenados)  # Saída: [1, 2, 3, 4]
```

```python
# o método copy() é usado para criar uma cópia superficial (shallow copy) de um vetor.

numeros = [1, 2, 3]
numeros_copia = numeros.copy()
print(numeros_copia)  # Saída: [1, 2, 3]
```

```python
# o operador de fatiamento permite extrair uma parte específica de um vetor.

numeros = [1, 2, 3, 4, 5]
subvetor = numeros[1:4]  # Extrai do índice 1 (inclusive) ao 4 (exclusive)
print(subvetor)  # Saída: [2, 3, 4]

```

```python
# a função enumerate() é usada para iterar sobre um vetor, fornecendo o índice e o valor correspondente de cada elemento.

frutas = ["maçã", "banana", "laranja"]
for indice, fruta in enumerate(frutas):
    print(indice, fruta)
```

```python
# a função zip() é usada para combinar elementos de dois ou mais vetores em pares.

letras = ['a', 'b', 'c']
numeros = [1, 2, 3]
combinado = zip(letras, numeros)
print(list(combinado))  # Saída: [('a', 1), ('b', 2), ('c', 3)]
```

```python
# o método clear() é usado para remover todos os elementos de um vetor, deixando-o vazio.

numeros = [1, 2, 3, 4, 5]
numeros.clear()
print(numeros)  # Saída: []
```

```python
# a função reversed() retorna um iterador reverso para percorrer os elementos de um vetor.

numeros = [1, 2, 3, 4, 5]
for numero in reversed(numeros):
    print(numero)
```

```python
# o operador not in é usado para verificar se um elemento não está presente no vetor.

frutas = ["maçã", "banana", "laranja"]
print("abacaxi" not in frutas)  # Saída: True
print("banana" not in frutas)  # Saída: False
```

```python
# a função sum() retorna a soma de todos os elementos de um vetor numérico.

numeros = [1, 2, 3, 4, 5]
soma = sum(numeros)
print(soma)  # Saída: 15
```

```python
# as funções max() e min() retornam o maior e o menor valor de um vetor, respectivamente.

numeros = [10, 5, 8, 3, 6]
maior = max(numeros)
menor = min(numeros)
print(maior)  # Saída: 10
print(menor)  # Saída: 3
```

```python
# as funções any() e all() verificam se pelo menos um ou todos os elementos de um vetor, respectivamente, atendem a uma condição.

numeros = [2, 4, 6, 8, 10]
todos_pares = all(num % 2 == 0 for num in numeros)
algum_impar = any(num % 2 != 0 for num in numeros)
print(todos_pares)  # Saída: True
print(algum_impar)  # Saída: False
```

```python
# a função map() permite aplicar uma função a cada elemento de um vetor e retorna um iterador com os resultados.

numeros = [1, 2, 3, 4, 5]
quadrados = map(lambda x: x ** 2, numeros)
print(list(quadrados))  # Saída: [1, 4, 9, 16, 25]
```

```python
# a função filter() permite filtrar os elementos de um vetor com base em uma condição especificada por uma função.

numeros = [1, 2, 3, 4, 5]
pares = filter(lambda x: x % 2 == 0, numeros)
print(list(pares))  # Saída: [2, 4]
```

# Topico 06: Matrizes

uma matriz é uma estrutura de dados bidimensional que pode ser representada como uma lista de listas. Cada lista interna representa uma linha da matriz, e os elementos individuais da matriz podem ser acessados usando a notação de índice.

Além disso, você pode realizar várias operações matriciais usando bibliotecas Python como **NumPy** ou **SciPy**, que fornecem funções e métodos avançados para manipulação de matrizes.

```python
# Criando uma matriz 2x3
matriz = [[1, 2, 3],
          [4, 5, 6]]

# Acessando elementos da matriz
elemento = matriz[0][1]  # Acessa o elemento da linha 0, coluna 1 (valor: 2)
print(elemento)  # Saída: 2

# Modificando elementos da matriz
matriz[1][2] = 7  # Modifica o elemento da linha 1, coluna 2 para 7
print(matriz)  # Saída: [[1, 2, 3], [4, 5, 7]]
```

```python
# Criando uma matriz 3x3
matriz = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]

# Percorrendo os elementos da matriz
for linha in matriz:
    for elemento in linha:
        print(elemento)
```

# Topico 09: Arquivos

### Entrada de arquivos

Em Python, você pode abrir e ler arquivos para obter informações. A função **open()** é usada para abrir um arquivo e o método **read()** é usado para ler o conteúdo do arquivo. Ao abrir um arquivo em modo de escrita, você pode usar o método **write()** para gravar conteúdo no arquivo. O método write() aceita uma string como argumento e grava essa string no arquivo.

- 'r': Modo de leitura (read). Esse é o modo padrão e permite ler o conteúdo de um arquivo. Se o arquivo não existir, ocorrerá um erro.
- 'w': Modo de escrita (write). Esse modo cria um novo arquivo ou substitui o conteúdo de um arquivo existente. Se o arquivo não existir, ele será criado. Se existir, o conteúdo anterior será apagado.
- 'a': Modo de anexar (append). Esse modo permite adicionar conteúdo a um arquivo existente. Se o arquivo não existir, ele será criado.
- 'x': Modo de criação exclusiva (exclusive creation). Esse modo cria um novo arquivo, mas gera um erro se o arquivo já existir.
- 'b': Modo binário (binary). Esse modo é usado para arquivos binários, como imagens, vídeos ou qualquer arquivo não textual.
- 't': Modo de texto (text). Esse é o modo padrão e é usado para arquivos de texto.

```python
# Entrada de arquivo
arquivo = open("dados.txt", "r")
conteudo = arquivo.read()
print(conteudo)
arquivo.close()
```

```python
# Gravando em um arquivo
arquivo = open('nome_do_arquivo.txt', 'w')
arquivo.write('Olá, mundo!')
arquivo.close()
```

```python
# anexar conteúdo a um arquivo existente
with open('arquivo_existente.txt', 'a') as arquivo:
    arquivo.write('Texto adicional no final do arquivo.\n')
    arquivo.write('Mais uma linha.')
```

```python
# criação exclusiva de um novo arquivo
try:
    with open('novo_arquivo.txt', 'x') as arquivo:
        arquivo.write('Este é um novo arquivo criado exclusivamente.')
except FileExistsError:
    print('O arquivo já existe.')
```

### O JSON


JSON (JavaScript Object Notation) é um formato popular para armazenar e trocar dados estruturados. Em Python, você pode trabalhar com JSON usando o módulo embutido chamado json.

```python
import json

dados = {
    'nome': 'João',
    'idade': 25,
    'cidade': 'São Paulo'
}

json_string = json.dumps(dados)
print(json_string)
# Saída: {"nome": "João", "idade": 25, "cidade": "São Paulo"}
```

```python
import json

json_string = '{"nome": "João", "idade": 25, "cidade": "São Paulo"}'

dados = json.loads(json_string)
print(dados)
# Saída: {'nome': 'João', 'idade': 25, 'cidade': 'São Paulo'}
```

```python
import json

# Gravando em um arquivo JSON
dados = {
    'nome': 'João',
    'idade': 25,
    'cidade': 'São Paulo'
}

with open('dados.json', 'w') as arquivo:
    json.dump(dados, arquivo)

# Lendo de um arquivo JSON
with open('dados.json', 'r') as arquivo:
    dados_lidos = json.load(arquivo)

print(dados_lidos)
# Saída: {'nome': 'João', 'idade': 25, 'cidade': 'São Paulo'}
```

# Topico 10: POO (Programação Orientada a Objetos)

é um paradigma de programação que organiza o código em estruturas chamadas de objetos. Em Python, você pode usar POO para criar classes e objetos, que são instâncias dessas classes.

### Classes e Objetos

Uma classe é uma estrutura que define um **conjunto de propriedades (atributos)** e **comportamentos (métodos)** que os objetos desse tipo podem ter. Uma vez definida a classe, você pode criar objetos a partir dela, que são instâncias da classe.

image.png

```python
class Pessoa:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

    def saudacao(self):
        print(f"Olá, meu nome é {self.nome} e eu tenho {self.idade} anos.")

# Criação de objetos da classe Pessoa
pessoa1 = Pessoa("João", 25)
pessoa2 = Pessoa("Maria", 30)

pessoa1.saudacao()  # Saída: Olá, meu nome é João e eu tenho 25 anos.
pessoa2.saudacao()  # Saída: Olá, meu nome é Maria e eu tenho 30 anos.
```

### Atributos

Os atributos são variáveis associadas a um objeto. Eles armazenam informações sobre o estado do objeto. Os atributos podem ser definidos no método especial ' __ init __ ' da classe

```python
class Pessoa:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

pessoa = Pessoa("João", 25)
print(pessoa.nome)  # Saída: João
print(pessoa.idade)  # Saída: 25
```

### Métodos

Os métodos são funções definidas dentro de uma classe que realizam operações específicas nos objetos da classe. Eles podem acessar e manipular os atributos do objeto.

```python
class Pessoa:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

    def saudacao(self):
        print(f"Olá, meu nome é {self.nome} e eu tenho {self.idade} anos.")

pessoa = Pessoa("João", 25)
pessoa.saudacao()  # Saída: Olá, meu nome é João e eu tenho 25 anos.
```

### Herança

A herança permite criar classes que herdam atributos e métodos de outra classe. A classe que é herdada é chamada de classe pai ou superclasse, e a classe que herda é chamada de classe filha ou subclasse.

```python
class Animal:
    def __init__(self, nome):
        self.nome = nome

    def som(self):
        pass

class Cachorro(Animal):
    def som(self):
        print("Au au!")

cachorro = Cachorro("Rex")
cachorro.som()  # Saída: Au au!
```

### Encapsulamento

é um conceito importante da programação orientada a objetos que permite **controlar o acesso aos atributos e métodos** de uma classe. Em Python, o encapsulamento é alcançado através do uso de convenções de nomenclatura e modificadores de acesso.



*   **Público** Os membros públicos de uma classe podem ser acessados de qualquer lugar do código, tanto dentro quanto fora da classe. Por padrão, todos os atributos e métodos são públicos em Python.

*   **Protegido** Os membros protegidos de uma classe têm uma convenção de nomenclatura específica e podem ser acessados apenas dentro da classe e suas subclasses. Os membros protegidos começam com um único sublinhado **_**

*   **Privado** Os membros privados de uma classe também têm uma convenção de nomenclatura específica e são acessados apenas dentro da própria classe. Os membros privados começam com dois sublinhados **_ _**



```python
# Atributo publico
class Exemplo:
    def __init__(self):
        self.atributo_publico = 10

    def metodo_publico(self):
        return "Método público"
```

```python
# Atributo protegido

class Exemplo:
    def __init__(self):
        self._atributo_protegido = 20

    def _metodo_protegido(self):
        return "Método protegido"
```

```python
# Atributo privado

class Exemplo:
    def __init__(self):
        self.__atributo_privado = 30

    def __metodo_privado(self):
        return "Método privado"
```

### polimorfismo

é um conceito importante da programação orientada a objetos que permite que objetos de diferentes classes sejam tratados de maneira uniforme, ou seja, é a capacidade de **um objeto assumir diferentes formas ou comportamentos.**

o polimorfismo é alcançado através do uso de herança e métodos comuns em diferentes classes. Quando várias classes têm um método com o mesmo nome, é possível chamar esse método em objetos dessas classes diferentes, e cada objeto irá executar sua própria implementação desse método.

```python
class Animal:
    def emitir_som(self):
        pass

class Cachorro(Animal):
    def emitir_som(self):
        print("O cachorro faz 'Au Au!'")

class Gato(Animal):
    def emitir_som(self):
        print("O gato faz 'Miau Miau!'")

cachorro = Cachorro()
gato = Gato()

cachorro.emitir_som()  # Saída: O cachorro faz 'Au Au!'
gato.emitir_som()  # Saída: O gato faz 'Miau Miau!'

```

O polimorfismo também pode ser utilizado em situações onde temos uma lista de objetos de diferentes classes, e podemos percorrer essa lista e chamar o mesmo método em cada objeto, independentemente da sua classe:

```python
animais = [cachorro, gato]

for animal in animais:
    animal.emitir_som()
```
