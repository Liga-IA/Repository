# Topico 09: Arquivos

### Entrada de arquivos

Em Python, você pode abrir e ler arquivos para obter informações. A função **open()** é usada para abrir um arquivo e o método **read()** é usado para ler o conteúdo do arquivo. Ao abrir um arquivo em modo de escrita, você pode usar o método **write()** para gravar conteúdo no arquivo. O método write() aceita uma string como argumento e grava essa string no arquivo.

- 'r': Modo de leitura (read). Esse é o modo padrão e permite ler o conteúdo de um arquivo. Se o arquivo não existir, ocorrerá um erro.
- 'w': Modo de escrita (write). Esse modo cria um novo arquivo ou substitui o conteúdo de um arquivo existente. Se o arquivo não existir, ele será criado. Se existir, o conteúdo anterior será apagado.
- 'a': Modo de anexar (append). Esse modo permite adicionar conteúdo a um arquivo existente. Se o arquivo não existir, ele será criado.
- 'x': Modo de criação exclusiva (exclusive creation). Esse modo cria um novo arquivo, mas gera um erro se o arquivo já existir.
- 'b': Modo binário (binary). Esse modo é usado para arquivos binários, como imagens, vídeos ou qualquer arquivo não textual.
- 't': Modo de texto (text). Esse é o modo padrão e é usado para arquivos de texto.
